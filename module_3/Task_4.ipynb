```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load a small dataset (Iris)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
data = pd.read_csv(url)
```

```python
# 1. BASIC LINE PLOT (Matplotlib)
plt.figure(figsize=(8, 5))
plt.plot(data.index, data['sepal_length'], color='blue', linewidth=2)
plt.title('Line Plot of Sepal Length')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.grid(True)
plt.show()
```

```python
# 2. SCATTER PLOT (Matplotlib)
# -----------------------------
plt.figure(figsize=(8, 5))
plt.scatter(data['sepal_length'], data['petal_length'], color='green')
plt.title('Scatter Plot of Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# 3. HISTOGRAM (Seaborn)
# -----------------------------
plt.figure(figsize=(8, 5))
sns.histplot(data['sepal_width'], bins=20, kde=True, color='purple')
plt.title('Histogram of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```
